**정리**
- label: 긍정 / 부정 판단
- score: 신뢰도 (0~1, 1에 가까울수록 확실함)

### 결과 정리 (1)

```python
positive_count=0 #초기값: 0
negative_count=0
```

**📌zip**
- 두 리스트를 짝 지어주는 함수
- review&result 같이 출력하게 해 줌

```python
for i, (review, result) in enumerate(zip(reviews, output)):
    label=result['label']   #긍정/부정
    score=result['score']   #신뢰도

    if label=='positive':
        print(f"[{i}]⭕ 긍정")
        positive_count+=1   #긍정
    else:
        print(f"[{i}] ❌ 부정")
        negative_count+=1   #부정
```

```python
for i, (review, result) in enumerate(zip(reviews, output)):
    label=result['label']   #긍정/부정
    score=result['score']   #신뢰도

    if label=='positive':
        print(f"[{i}]⭕ 긍정 ({score:.0%}) | {review[:20]}")   #reiviews =! reivew임
        positive_count+=1   #긍정
    else:
        print(f"[{i}] ❌ 부정 ({score:.0%}) | {review[:20]}")
        negative_count+=1   #부정
```

```python
print(f"\n ⭕긍정: {positive_count}개")
print(f" ❌부정: {negative_count}개")
```
